```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
match_data = pd.read_csv('WorldCupMatches.csv')
df = pd.DataFrame(match_data)
```

```python
#1. NO. OF ROWS AND COLUMNS
rows, columns = df.shape
print(f"No. of rows: {rows}")
print(f"No. of columns: {columns}")
```

```python
#PRINT EACH COLUMN NAME
df.columns
```

```python
#PRINT SPECIFIC RANGE USING SLICING
print(df.iloc[0:20, 0:5])
```

```python
#SUMMARIZATION STATISTICS
df.describe()
```

```python
#CHECKING MISSING VALUES
df.isnull()
#isnull() is use for find the missing values in a dataset
#isnull().sum() is used for couting the number of null values in a column
```

```python
#REPLACING THE COLUMN "HOME TEAM GOALS" NULL VALUES WITH ITS MEAN VALUES
df.fillna({'Home Team Goals':df['Home Team Goals'].mean()}, inplace=True)
df
```

```python
#REPLACING THE COLUMN "ATTENDENCE" NULL VALUES WITH MEDIAN VALUES
df.fillna({'Attendance':df['Attendance'].median()}, inplace=True)
df
```

```python
#PAIR PLOTTING
sns.pairplot(data=match_data, hue='City')
plt.title('PAIR PLOTTING OF EACH CITY')
plt.show()
```

```python
#BOX PLOTTING
sns.boxplot(data=match_data, x='City', y='Attendance')
plt.title('Attendance by City')
```

```python
#BAR PLOTTING
sns.barplot(data=match_data, x='City', y='Attendance')
plt.title('BAR PLOT FOR ATTENDANCE BY CITY')
```

```python
#HISTOGRAM
sns.histplot(data=match_data, x='City')
plt.title('HISTOGRAM FOR EACH CITY')
```

```python
#DISTRIBUTION PLOT
sns.displot(df['City'], kde=True)
plt.title("DISTRIBUTION PLOT FOR CITY")
```

```python
#SCATTER PLOTTING
sns.scatterplot(data=match_data, x='City', y='Attendance')
plt.title('SCATTER PLOT Attendance by City')
```

```python
#HEATMAP
columns = ['Year', 'Home Team Goals', 'Away Team Goals', 'Attendance', 'Half-time Home Goals','Half-time Away Goals']
corr_data = df[columns].corr()
sns.heatmap(data=corr_data, annot=True, cmap='viridis', linewidths=0.5)
plt.title('Correlation Heat Map')
plt.show()
```
